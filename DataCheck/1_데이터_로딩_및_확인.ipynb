## 파일 시작부분에 불필요내용이 있고, 또한 헤더가 없는 xyz.txt 파일 데이터 로딩

```python
# xyz.txt 파일 시작부분에 불필요한 내용이 들어 있는 경우 skiprows 옵션사용
xyz = pd.read_csv('xyz.txt', sep='\t', skiprows = ([0,1]), skipinitialspace=True)
```

```python
# xyz.txt 파일을 로딩시 첫번째 레코드가 헤더로 잘못 인식됨
xyz.head()
```

```python
# xyz.txt 파일에 헤더도 없으므로 이를 header 옵션(0 또는 None)을 통해 로딩시 알려줌
xyz = pd.read_csv('xyz.txt', sep='\t', skiprows = ([0,1]), header = None, skipinitialspace=True)
```

```python
# xyz.txt 파일의 정상 로딩결과 --> xyz 데이터프레임의 헤더에 각 컬럼변수병이 숫자로 기본설정됨
xyz.head()
```

```python
# xyz데이터프레임의 숫자컬럼변수명을 적정한 변수컬럼명으로 변환시킴
xyz.columns = ['gender', 'job', 'age', 'position', 'address', 'total', 'check', 'price', 'survey']
```

```python
# xyz 데이터프레임의 변경된 컬럼변수명 확인
xyz.head()
```

```python
# xyz데이터프레임의 변수컬럼명 조회
xyz.columns
```

```python
# xyz데이터프레임의 특정변수컬럼의 이름변경
xyz.rename(columns={'total':'amount'}, inplace=True)
```

```python
# xyz 데이터프레임의 변경된 컬럼변수명 확인
xyz.head()
```

```python
# xyz데이터프레임 중 여러 개의 변수컬럼 이름 동시변경
xyz.rename(columns={'job':'occupation', 'price':'payment'}, inplace=True)
```

```python
# xyz 데이터프레임의 변경된 컬럼변수명 확인
xyz.head()
```

## 엑셀파일 ex.xlsx 데이터 로딩

```python
# ex.xlsx 엑셀파일을 로딩하기
ex = pd.read_excel('ex.xlsx')
```

```python
# 로딩된 ex.xlsx 엑셀파일의 내용을 출력해보면 첫번째 시트인 aaa시트가 로딩된 것을 알 수 있음 
ex.head()
```

```python
# ex.xlsx 파일의 두번째 시트인 bbb시트를 로딩하는 방법 ==> sheetname=1이라고 해도 됨
bbb = pd.read_excel('ex.xlsx', header = None, sheet_name = 'bbb')
```

```python
# 두번째 시트 bbb가 제대로 로딩되었는지 확인 
# 원본 bbb시트에 헤더가 없는 상태라 숫자컬럼이 임시로 반영됨
bbb.head()
```

```python
# bbb 데이터프레임의 숫자컬럼을 변수컬럼으로 변경
bbb.columns = ['gender', 'job', 'age', 'position', 'address', 'total', 'check', 'price', 'survey']
bbb.head()
```
